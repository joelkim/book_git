# git이란

- git은 분산 버전 관리를 위한 커맨드라인 도구이다.

## git은 커맨드라인 도구

- git은 시스템이 아닌 커맨드라인 도구이다.
- git이라는 이름의 프로세스가 존재하지 않고 git 명령을 입력하여 실행할 때마다 특정한 작업을 수행하고 종료된다.

## 실습 1: git 명령어

- shell을 열고 다음 명령어를 실행해보자.

```bash
git || true
```

- git 명령 다음에 서브 명령어 또는 옵션을 입력하라는 메세지가 길게 출력된다.
- `--version` 옵션을 사용하여 git의 버전을 확인해보자.

```bash
git --version
```
